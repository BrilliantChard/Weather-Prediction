# **Machine Learning**

```python
# Determine the Target Variable

core_weather_data['target'] = core_weather_data.shift(-1)['Temp_max']
core_weather_data
```

```python
core_weather_data = core_weather_data.iloc[:-1, :].copy()
core_weather_data
```

```python
from sklearn.linear_model import  Ridge
reg = Ridge(alpha=0.1)
```

```python
core_weather_data
```

```python
predictors = ['Snow_depth', 'Temp_avg', 'Temp_max', 'Temp_min']
```

```python
train = core_weather_data.loc[: '2022-12-31']
test = core_weather_data.loc['2023-01-01' : ]
```

```python
reg.fit(train[predictors], train['target'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
print(test['target'].shape)
print(predictions.shape)
```

```python
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error
```

```python
mae = mean_absolute_error(test['target'], predictions)

print(f'Mean Absolute Error: {mae:.3f}')
```

```python
plt.plot(test, predictions)
plt.show()
```

# **Evaluating the Model**

```python
combined = pd.concat([test['target'], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ['Actual', 'Predictions']
combined
```
